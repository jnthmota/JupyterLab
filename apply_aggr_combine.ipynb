## Min-Max Scalling

```python
ages = df['age']
ages
```

```python
ages.min()
```

```python
ages.max()
```

```python
def min_max_scale_ages(val):
    min_age = 17
    max_age = 90
    return (val-min_age)/(max_age-min_age)

min_max_ages = ages.apply(min_max_scale_ages)

min_max_ages
```

---
Vemos que `.apply()` recebe uma funÃ§Ã£o aplicÃ¡vel em cada valor da Series, individualmente.

Trazendo pro exemplo, cada linha de ages Ã© um `int`, portanto a funÃ§Ã£o de dentro do apply deve ser uma funÃ§Ã£o que espera receber 1 `int`.
Python Ã© python, entÃ£o se ages fosse uma Series de `floats` tudo funcionaria tambÃ©m, mas Ã© bom ter cuidado com isso.

Por exemplo, o que aconteceria se algum dos valores fosse `NaN`?
E se fosse outra funÃ§Ã£o, uma que esperasse strings, e de repente recebe um `NaN`, ou atÃ© mesmo um `int`, jÃ¡ que 'object' pode ser qualquer `type`?

---
Bom, existe um apply tambÃ©m para `DataFrames`.
Qual a expectativa que temos a respeito de que tipo de funÃ§Ã£o pode ser usada?

```python
df_ints = df.select_dtypes(include='int64')

df_ints.head()
```

```python
def tell_me_type(x):
    print("x is type:", type(x))

df_ints.apply(tell_me_type)
```

```python
def min_max_scale_series(ser):
    min_val = ser.min()
    max_val = ser.max()

    def internal_scaler(val):
        return (val-min_val)/(max_val-min_val)
    
    return ser.apply(internal_scaler)

df_ints.head().apply(min_max_scale_series, axis='index')
```

```python
df_ints.head().apply(min_max_scale_series, axis='columns')
```

```python
def nice_row_classifier(row):
    if (row['age'] * row['capital-gain']) - row.mean() > (row['fnlwgt'] / row['education-num']):
        return "Nice"
    else:
        return "Cool"

df_ints.apply(nice_row_classifier, axis='columns')
```

```python
df_strs
```

```python
def call_lower(s):
    if isinstance(s, str):
        return s.lower()
    else:
        return s

df_strs.applymap(call_lower)
```

---
EntÃ£o, entre `.apply()`  e `.applymap()` podemos aplicar funÃ§Ãµes em DataFrames como um todo.
Quer seja elemento a elemento, quer seja linha por linha ou coluna por coluna.

Ainda assim, podem existir situaÃ§Ãµes em que se deseja iterar pelo DataFrame.

JÃ¡ vimos que coluna a coluna podemos simplesmente usar um laÃ§o for pelo df e o elemento que Ã© iterado sÃ£o os nomes das colunas, nos permitindo fazer esse acesso facilmente.
Ainda assim, ainda existe `DataFrame.items()` para iterar por colunas (fica como dica para vocÃªs pesquisarem como usar) 

JÃ¡ para iterar dpor linhas temos 2 boas opÃ§Ãµes: `DataFrame.iterrows()` e `DataFrame.itertuples()`

```python
for index, row in df_ints.head().iterrows():

    print(index, type(row), row['age'], row.mean())
```

```python
for row in df_ints.head().itertuples():

    print(row, type(row))
    print(row.Index, row.age, row._3)
```

## Agrupando Dados

```python
relationship_group = df.groupby('relationship')
```

```python
relationship_group.get_group('Married') # resulta no mesmo que df.loc[ df['relationship'] == 'Married' ]
```

```python
def my_fst(s):
    return s.iloc[0]

agg_lst = ['mean', my_fst]

relationship_group.agg(agg_lst)
```

```python
agg_dict = {
    'age': agg_lst,
    'fnlwgt': 'mean',
    'workclass': my_fst
}

relationship_group.agg(agg_dict)
```
