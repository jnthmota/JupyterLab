# Aplicando Funções em Pandas

Vamos começar alterando valores de uma coluna específica.
Lembrando que quando trabalhamos com uma coluna por vez, estamos usando `Series`. 

```python
pd.crosstab(df['relationship'], df['sex'])
```

Já que já temos informação de 'sex' em coluna separada e as exceções são espúrias, vamos remover a distinção entre 'Husband' e 'Wife' e trocar esses valores por 'Married'.

```python
to_map = {'Husband': 'Married', 'Wife': 'Married'}

df['relationship'].map(to_map)
```

```python
to_map = {val: val for val in df['relationship'].unique()}
to_map['Husband'] = 'Married'
to_map['Wife'] = 'Married'

df['relationship'].map(to_map)
```

```python
to_replace = {'Husband': 'Married', 'Wife': 'Married'}

df['relationship'].replace(to_replace)
```

```python
df['relationship'].unique()
```

```python
df['relationship'].replace(to_replace, inplace=True)
```

```python
df['relationship'].unique()
```

---
Existe uma alternativa mais razoável aqui... o método `.replace` não precisa dar essa volta.

O ponto importante aqui é que `.map` TEM que ser aplicado em cada valor da Series.
Quando o valor não pode ser encontrado... ele é `NA` (not available)
É como se o dicionário fosse uma função que especifica na unha que resultado atribuir a cada valor (matematicamente uma função de fato).

Ok. Mas e se eu precisar de fato computar algo pra gerar o resultado?
Aí vamos precisar de outro método.
Vamos trabalhar isso através de um exemplo... 

## Min-Max Scalling

```python
ages = df['age']
ages
```

```python
ages.min()
```

```python
ages.max()
```

```python
def min_max_scale_ages(val):
    min_age = 17
    max_age = 90
    return (val-min_age)/(max_age-min_age)

min_max_ages = ages.apply(min_max_scale_ages)

min_max_ages
```

---
Vemos que `.apply()` recebe uma função aplicável em cada valor da Series, individualmente.

Trazendo pro exemplo, cada linha de ages é um `int`, portanto a função de dentro do apply deve ser uma função que espera receber 1 `int`.
Python é python, então se ages fosse uma Series de `floats` tudo funcionaria também, mas é bom ter cuidado com isso.

Por exemplo, o que aconteceria se algum dos valores fosse `NaN`?
E se fosse outra função, uma que esperasse strings, e de repente recebe um `NaN`, ou até mesmo um `int`, já que 'object' pode ser qualquer `type`?

---
Bom, existe um apply também para `DataFrames`.
Qual a expectativa que temos a respeito de que tipo de função pode ser usada?

```python
df_ints = df.select_dtypes(include='int64')

df_ints.head()
```

```python
def tell_me_type(x):
    print("x is type:", type(x))

df_ints.apply(tell_me_type)
```

```python
def min_max_scale_series(ser):
    min_val = ser.min()
    max_val = ser.max()

    def internal_scaler(val):
        return (val-min_val)/(max_val-min_val)
    
    return ser.apply(internal_scaler)

df_ints.head().apply(min_max_scale_series, axis='index')
```

```python
df_ints.head().apply(min_max_scale_series, axis='columns')
```

```python
def nice_row_classifier(row):
    if (row['age'] * row['capital-gain']) - row.mean() > (row['fnlwgt'] / row['education-num']):
        return "Nice"
    else:
        return "Cool"

df_ints.apply(nice_row_classifier, axis='columns')
```

```python
df_strs
```

```python
def call_lower(s):
    if isinstance(s, str):
        return s.lower()
    else:
        return s

df_strs.applymap(call_lower)
```

---
Então, entre `.apply()`  e `.applymap()` podemos aplicar funções em DataFrames como um todo.
Quer seja elemento a elemento, quer seja linha por linha ou coluna por coluna.

Ainda assim, podem existir situações em que se deseja iterar pelo DataFrame.

Já vimos que coluna a coluna podemos simplesmente usar um laço for pelo df e o elemento que é iterado são os nomes das colunas, nos permitindo fazer esse acesso facilmente.
Ainda assim, ainda existe `DataFrame.items()` para iterar por colunas (fica como dica para vocês pesquisarem como usar) 

Já para iterar dpor linhas temos 2 boas opções: `DataFrame.iterrows()` e `DataFrame.itertuples()`

```python
for index, row in df_ints.head().iterrows():

    print(index, type(row), row['age'], row.mean())
```

```python
for row in df_ints.head().itertuples():

    print(row, type(row))
    print(row.Index, row.age, row._3)
```

## Agrupando Dados

```python
relationship_group = df.groupby('relationship')
```

```python
type(relationship_group)
```

```python
relationship_group.groups
```

```python
relationship_group.get_group('Married') # resulta no mesmo que df.loc[ df['relationship'] == 'Married' ]
```

```python
relationship_group.mean()
```

```python
grouped_means = df.groupby(by=['sex', 'relationship']).mean()

grouped_means
```

```python
grouped_means.loc['Female']
```

```python
type(grouped_means.loc['Female'])
```

```python
(grouped_means.loc['Female']).iloc[1, 2]
```

```python
grouped_means.loc['Female', 'Married']['age']
```

```python
def my_fst(s):
    return s.iloc[0]

agg_lst = ['mean', my_fst]

relationship_group.agg(agg_lst)
```

```python
agg_dict = {
    'age': agg_lst,
    'fnlwgt': 'mean',
    'workclass': my_fst
}

relationship_group.agg(agg_dict)
```
