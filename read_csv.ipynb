Já foi possível resolver problemas de excesso de espaços na leitura.

Uma vez que o DataFrame foi instanciado, podemos executar esse *strip* a posteriori, mas é mais eficiente fazer já na leitura, quando suficientemente possível e conveniente.

```python
for col in df.select_dtypes(include='object'):
    
    print(col, any(df[col].str.startswith('?')))
```

```python
for col in df.select_dtypes(include='object'):
    
    print(col, any(df[col].isna()))
```

---
A documentação afirma que valores desconhecidos (missings) são marcados por uma string `'?'`.

Podemos ver que ? foram lidos como strings, mas podemos sinalizar a convenção na leitura.

```python
df = pd.read_csv('data', header=None, skipinitialspace=True, na_values='?')

for col in df.select_dtypes(include='object'):
    
    print(col, any(df[col].isna()))
```

```python
for col in df.select_dtypes(include='object'):
    
    print(col, any(df[col].str.startswith('?')))
```

---
Observem que ? foi sinalizado como missing, mas não houve alteração nas strings em si

```python
df
```

```python
df = pd.read_csv('data', header=None, skipinitialspace=True, na_values='?', true_values=['>50K'], false_values=['<=50K'])

df
```

Observem que, dessa vez, os valores de fato foram convertidos
