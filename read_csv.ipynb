----
JÃ¡ foi possÃ­vel resolver problemas de excesso de espaÃ§os na leitura.

Uma vez que o DataFrame foi instanciado, podemos executar esse *strip* a posteriori, mas Ã© mais eficiente fazer jÃ¡ na leitura, quando suficientemente possÃ­vel e conveniente.

```python
for col in df.select_dtypes(include='object'):
    
    print(col, any(df[col].str.startswith('?')))
```

```python
for col in df.select_dtypes(include='object'):
    
    print(col, any(df[col].isna()))
```

---
A documentaÃ§Ã£o afirma que valores desconhecidos (missings) sÃ£o marcados por uma string `'?'`.

Podemos ver que ? foram lidos como strings, mas podemos sinalizar a convenÃ§Ã£o na leitura.

```python
df = pd.read_csv('data', header=None, skipinitialspace=True, na_values='?')

for col in df.select_dtypes(include='object'):
    
    print(col, any(df[col].isna()))
```

```python
for col in df.select_dtypes(include='object'):
    
    print(col, any(df[col].str.startswith('?')))
```

---
Observem que ? foi sinalizado como missing, mas nÃ£o houve alteraÃ§Ã£o nas strings em si
