---
Uma das funções padrão a serem implementadas para classes em Python é a função `len`.
Para strings, listas, tuplas, conjuntos, dicionários e outras classes que implementam coleções, esse costuma ser o comportamento.

No entanto, em DataFrames, é comum querermos saber quantos são os valores válidos além de saber quantos são os valores.
Para isso, temos o `.count()`, que apresenta a contagem de valores não missing existentes na estrutura.

Existe tanto a versão de `.count()` para `DataFrame` quanto para `Series`.
A primeira retorna os valores que seriam retornados por cada coluna.

Para Series, além disso, temos `.value_counts()`.
Esse método retorna uma Series cujos índices são os valores únicos encontrados na Series passada e os valores são a contagem desses valores

Passando agora para transformações, já vimos que muitas operações algébricas conseguem se aproveitar dos operadores aritméticos e lógicos.

Então vamos separar as colunas de strings e trabalhar algumas transformaçẽos com elas...

````python
df_strs = df.select_dtypes(exclude='number')
````

````python
df_strs
````

````python
df_strs['sex'].value_counts()
````

````python
males_series = (df_strs['sex'] == 'Male')

males_series.value_counts()
````

````python
df_strs['sex'].unique()
````

````python
for val in df_strs['sex'].unique():
    print(val)
````

````python
for val in df_strs['sex'].unique():
    val_series = (df_strs['sex'] == val)
    
    print(val_series.value_counts())
````

````python
for val in df_strs['sex'].unique():
    df_strs.loc[:, val] = (df_strs['sex'] == val)

df_strs
````

````python
df_strs = df.select_dtypes(include='object').copy()

for col in df_strs:
    for val in df_strs[col].unique():
        df_strs.loc[:, val] = (df_strs[col] == val)

df_strs.head()
````
