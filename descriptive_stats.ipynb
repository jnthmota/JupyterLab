Uma das fun√ß√µes padr√£o a serem implementadas para classes em Python √© a fun√ß√£o `len`.
Para strings, listas, tuplas, conjuntos, dicion√°rios e outras classes que implementam cole√ß√µes, esse costuma ser o comportamento.

No entanto, em DataFrames, √© comum querermos saber quantos s√£o os valores v√°lidos al√©m de saber quantos s√£o os valores.
Para isso, temos o `.count()`, que apresenta a contagem de valores n√£o missing existentes na estrutura.

Existe tanto a vers√£o de `.count()` para `DataFrame` quanto para `Series`.
A primeira retorna os valores que seriam retornados por cada coluna.

Para Series, al√©m disso, temos `.value_counts()`.
Esse m√©todo retorna uma Series cujos √≠ndices s√£o os valores √∫nicos encontrados na Series passada e os valores s√£o a contagem desses valores

Passando agora para transforma√ß√µes, j√° vimos que muitas opera√ß√µes alg√©bricas conseguem se aproveitar dos operadores aritm√©ticos e l√≥gicos.

Ent√£o vamos separar as colunas de strings e trabalhar algumas transforma√ß·∫Ωos com elas...

````python
df_strs = df.select_dtypes(exclude='number')
````

````python
df_strs
````

````python
df_strs['sex'].value_counts()
````

````python
males_series = (df_strs['sex'] == 'Male')

males_series.value_counts()
````

````python
df_strs['sex'].unique()
````

````python
for val in df_strs['sex'].unique():
    print(val)
````

````python
for val in df_strs['sex'].unique():
    val_series = (df_strs['sex'] == val)
    
    print(val_series.value_counts())
````

````python
for val in df_strs['sex'].unique():
    df_strs.loc[:, val] = (df_strs['sex'] == val)

df_strs
````

````python
df_strs = df.select_dtypes(include='object').copy()

for col in df_strs:
    for val in df_strs[col].unique():
        df_strs.loc[:, val] = (df_strs[col] == val)

df_strs.head()
````
